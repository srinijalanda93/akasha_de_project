# Preprocessing - Handling the Duplicate Orders

```python
# We select only the columns that define a unique order
# and drop all duplicate rows
unique_orders_df = order_df[[
    'order_id', 
    'mobile_number', 
    'order_date_time', 
    'total_amount'
]].drop_duplicates()
```

```python

print("--- Unique Orders (First 5 Rows) ---")
display(unique_orders_df.head())

print(f"Went from {len(order_df)} total items to {len(unique_orders_df)} unique orders.")
```

# preprocessing - Cleaning Data Types

```python
# 1. Convert the date column from text (object) to a real datetime
unique_orders_df['order_date_time'] = pd.to_datetime(unique_orders_df['order_date_time'])

# 2. Convert total_amount to a number (it's good practice)
unique_orders_df['total_amount'] = pd.to_numeric(unique_orders_df['total_amount'])


print("--- Cleaned Data Info (Note the Dtypes) ---")
unique_orders_df.info()
```

```python
#finally let join the two dataset in order to get the region for each order they have common column is mobilenumber
```

```python
unique_orders_df
```

```python
customers_df     #  "right" table
```

```python
# This is the "join"
merged_df = pd.merge(
    unique_orders_df,   #  "left" table
    customers_df,       #  "right" table
    on='mobile_number', #  "key" to link them
    how='left'          # 'left' means "keep all orders, even if a customer is missing"
)

print("--- Fully Merged & Cleaned Data (First 5 Rows) ---")
display(merged_df.head())
```

```python
# regional Revenue
```

```python

print("KPI 3: Regional Revenue")

# 1. Group the table by the 'region' column
# 2. Select the 'total_amount' column
# 3. Sum() it all up
regional_revenue = merged_df.groupby('region')['total_amount'].sum().reset_index()

display(regional_revenue.sort_values(by='total_amount', ascending=False))
```
